```python
import pandas as pd
import numpy as np

class ReqNSatellite:
    
    def __init__(self, t, r = 6371300., L_ekv = 40075000,
                 H = 600000., M = 5.9722 * (10 ** 24), G = 6.67430 * (10 ** -11), l = 200000.):
        
        self.t = t # secunds
        self.L_ekv = L_ekv #m
        self.r = r #m
        self.H = H #m
        self.M = M #kg
        self.G = G # newtom * m ^ 2 / kg ^ 2
        self.l = l #m
        
    def calc_l_apostraf(self):
        
        return max(0, 2 * self.r * self.t * ((self.G * self.M / ((self.H + self.r) ** 3)) ** 0.5) - self.l)
    
    def calc_L(self):
        
        return 2 * self.t * ((self.G * self.M / (self.r + self.H)) ** 0.5)
    
    def count_n(self):
        n = -(-self.L_ekv / (self.l + self.calc_l_apostraf()) // 1) 
        return n
    
    def calc_dir_L(self):
        N = self.count_n()
        return 2 * (self.H + self.r) * np.sin(180 / N * 2 * np.pi / 360)

aa = ReqNSatellite(t = 10 * 60)
[aa.calc_l_apostraf() / 1000,
                   aa.calc_L() / 1000, aa.calc_dir_L() / 1000, aa.count_n(), aa.t / 60]

minn = 300
maxx = 3001

df = pd.DataFrame(columns=['расстояние между засечками', 'круговое расстояние между спутниками',
                          "прямое расстояние", 'количество спутников', "время"])

for time in range(minn, maxx, 300):
    
    obj = ReqNSatellite(t = time)
    df.loc[len(df.index)] = [obj.calc_l_apostraf() / 1000,
                   obj.calc_L() / 1000, obj.calc_dir_L() / 1000, obj.count_n(), time / 60]
df.to_excel("output.xlsx")
df
```
